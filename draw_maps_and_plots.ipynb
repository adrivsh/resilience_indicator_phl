# Some basic plots

```python
x = 100*df_with_results.risk
y = 100*df_with_results.resilience

plt.scatter(x,y)

plt.xlabel("Risk to welfare (%)")
plt.ylabel("Resilience (%)")

plt.xlim(0)
plt.ylim(0);
```

```python
x = 100*df_with_results.risk
y = 100*df_with_results.risk_to_assets

plt.scatter(y,x)

plt.ylabel("Risk to welfare (%)")
plt.xlabel("Risk to assets (%)")

plt.title("Lack of socio economic capacity can worsen risk")

plt.xlim(0,2)
plt.ylim(0)

plt.savefig("risk_assets_and_welfare.png")

plt.plot([0, 2],[0, 2], "--")

plt.savefig("risk_assets_and_welfare-1.png")

plt.annotate("",
            xy=(0.6, 0.6), xycoords='data',
            xytext=(0.6, 2.4), textcoords='data',
            arrowprops=dict(arrowstyle="<->",
                            connectionstyle="arc3", color="red"),
            )


plt.savefig("risk_assets_and_welfare-2.png")
```
